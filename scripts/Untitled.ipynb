```python
import pickle
import pandas as pd
```

```python
with open('data', "rb") as f:
        fdata = pickle.load(f)
```

```python
fdata
```

```python
data = pd.DataFrame([], columns=['car_x', 'car_y', 'act'])
```

```python
for d in fdata:
    try:
        nd = pd.DataFrame([[d['car']['x'], d['car']['y'], d['act']]], columns=['car_x', 'car_y', 'act'])
    except KeyError as e:
        nd = pd.DataFrame([[d['car']['x'], d['car']['y'], None]], columns=['car_x', 'car_y', 'act'])

    data.append(nd)
```
